```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create a simple DataFrame
data = pd.DataFrame({
    'EmployeeID': [1, 2, 3, 4, 5],
    'Age': [25, 30, 35, 40, 45],
    'Department': ['HR', 'Finance', 'IT', 'IT', 'Finance'],
    'Salary': [50000, 60000, 70000, 75000, 80000]
})

# Display the basic information about the DataFrame
print("Basic Information:")
print(data.info())

print("\nFirst Few Rows of the DataFrame:")
print(data.head())

# Descriptive statistics for numerical columns
print("\nDescriptive Statistics:")
print(data.describe())

# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())

# Plot histograms for numerical features
print("\nHistograms of Numerical Features:")
data[['Age', 'Salary']].hist(figsize=(10, 5), bins=10, color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Age and Salary')
plt.show()

# Boxplot for Salary by Department
print("\nBoxplot of Salary by Department:")
plt.figure(figsize=(8, 6))
sns.boxplot(x='Department', y='Salary', data=data, palette='Set2')
plt.title('Boxplot of Salary by Department')
plt.show()

# Pairplot to visualize relationships between numerical features
print("\nPairplot of Numerical Features:")
sns.pairplot(data[['Age', 'Salary']])
plt.suptitle('Pairplot of Age and Salary', y=1.02)
plt.show()
```
